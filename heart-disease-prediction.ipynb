Importing the libraries

```python
import numpy as np
import pandas as pd # used for creating dataframes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score # to evaluate our model
```

Data Collection and Data preprocessing

```python
# loading the csv data to a Pandas Datafram
heart_data=pd.read_csv('/kaggle/input/heart-disease-dataset/heart.csv')
```

```python
heart_data.head()
```

```python
# no of rows and column in dataset
heart_data.shape
```

```python
# getting some more onfo about data
heart_data.info()
```

```python
# checking for missing values ( another method)
heart_data.isnull().sum()
```

```python
# statistical measure about the data
heart_data.describe()
```

```python
# checking the destribution of target variable
# this will tell how many target variable are 0 and how many are 1
heart_data['target'].value_counts()
```

1 --> defective heart

0 --> healthy heart

Splitting the Features and Target

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

dropping the target column

whenever we are dropping a column we need to mention the axis =1 and if we want to drop a row mention axis =0

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training and Testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

0.2 -> represents 20% of the data

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
# Logistic Regression
model = LogisticRegression()
# training the logisticRegression model with training data
model.fit(X_train,Y_train) # this will find the relationship between the features in X_tarin and Y_train
```

Model Evaluation

```python
# Accuracy Score
# Accuracy on Training Data
X_train_pred=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,Y_train)
```

```python
print("Accuracy on Training data : ", training_data_accuracy)
```

0.85 means 85% accuracy means 85% of training data is predicted accurate

```python
# Accuracy on Test Data
X_test_pred=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,Y_test)
```

```python
print("Accuracy on Test data : ", test_data_accuracy)
```

Building  a Predictive System

```python
input_data=(41,0,1,130,204,0,0,172,0,1.4,2,0,2)

# change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
# print(input_data_reshaped)
prediction= model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
    print("The person does not have a Heart Disease")
else:
    print("The person has Heart Disease")
```
